## Условное математическое ожидание

### Задача 1

Итак, нам известно что $N_t - N_s \sim Pois(\lambda(t-s))$ и является незавсимой с величиной $N_s$. Тогда найдем $$E(N_t |N_s) = E(N_t - N_s + N_s| N_s) = E(N_t - N_s | N_s) + E(N_s | N_s) = N_s + E(N_t - N_s) = N_s + \lambda (t-s).$$ Здесь восрользовались линейностью УМО, а такаже свойствами, что если $\xi$ и $\eta$ независимы, то $E(\xi | \eta) = E(\xi)$, если же $\xi$ является $\eta$-измеримеримой случайной величиной, то $E(\xi | \eta) = \xi$.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import re
import time
import matplotlib.pyplot as plt

#считывание данных
data = pd.read_csv("6.csv.xls", header=None)
data = data.values
parametrs = re.findall(r'\b[0-9]+', data[0][0] + data[1][0] + data[2][0])
data = data[3:]
lmbda = 1/int(parametrs[0])
t0 = int(parametrs[1])
t = int(parametrs[2])
n = 0   #количество сломанных компьютеров к моменту времени s
print("Параметры: lambda = {}, t0 = {}, t = {}".format(lmbda, t0, t))  #Печатаем параметры
for s in range(0, t + 1, t0):
    while(n < data.size and float(data[n]) <= s):  #бежим по массиву данных сравнивая и смотрим какие серверы
        n += 1  #вышли из строя к моенту времени s, растет монотонно 
    prediction = n + lmbda * (t - s)
    print("s = {}, N_s = {}, prediction = {}".format(s, n, int(prediction))) #вывод данных
    time.sleep(t0)  #засыпаем на t0 секунд

#заметим, что здесь prediciton выводится в int, так как это величина вообще говоря целая по смылслу 
```

Итак, видим что с помощью УМО мы смогли достаточно точно спрогнозировать количество вышедших из строя серверов (так как изначально значение prediction совпадает практически с итоговым). При этом с ростом времени времени s, прогноз ведет себя не монотонно, в какие-то моменты времени он ухудшается, а в какие-то наооборот улучшается, так что нельзя сказать что с ростом времени s прогноз как-то улучшается. Конечное точное значение - 1000, когда $s = t$.
